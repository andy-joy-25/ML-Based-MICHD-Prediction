```python
def calc_MI(X,Y,bins):
 
    c_XY = np.histogram2d(X,Y,bins)[0]
    c_X = np.histogram(X,bins)[0]
    c_Y = np.histogram(Y,bins)[0]
 
    H_X = shan_entropy(c_X)
    H_Y = shan_entropy(c_Y)
    H_XY = shan_entropy(c_XY)
 
    MI = H_X + H_Y - H_XY
    return MI
 
def shan_entropy(c):
    c_normalized = c / float(np.sum(c))
    c_normalized = c_normalized[np.nonzero(c_normalized)]
    H = -sum(c_normalized * np.log2(c_normalized))  
    return H
 
"from https://stackoverflow.com/questions/20491028/optimal-way-to-compute-pairwise-mutual-information-using-numpy"
```

```python
BINS = [5, 100, 200, 1_000]
result = {}
fix, axes = plt.subplots(1, 2, figsize=(14, 4))
for bins in BINS:
    for i, col in enumerate(NAMES):
        mi = calc_MI(data[:,i], target, bins)
        result[col] = mi

    values = list(result.values())

    axes[0].plot(np.cumsum(sorted(values)[::-1]), label="Bins: {}".format(bins))
    axes[1].plot(sorted(values)[::-1], label="Bins: {}".format(bins))

axes[1].axvline(200, linestyle="dashed", c="black")
axes[1].legend()
axes[1].set_ylabel("Cumulative MI Score Captured")
axes[1].set_xlabel("Number of Features")

axes[0].axvline(200, linestyle="dashed", c="black")
axes[0].legend()
axes[0].set_ylabel("MI Scores across Features")
axes[0].set_xlabel("Number of Features")

plt.tight_layout()

```
